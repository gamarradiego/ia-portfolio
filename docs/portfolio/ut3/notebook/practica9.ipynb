```python
# Importar librerÃ­as necesarias
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, applications, optimizers, callbacks
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')

# ConfiguraciÃ³n de GPU
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
    print("ğŸ”§ GPU configurada correctamente")
else:
    print("ğŸ”§ Usando CPU")

# Seeds para reproducibilidad
tf.random.set_seed(42)
np.random.seed(42)

print("âœ… Entorno TensorFlow/Keras configurado correctamente")

# === PREPARAR DATASET CIFAR-10 ===

print("ğŸ“Š PREPARANDO DATASET CIFAR-10")
print("-" * 50)

# 1. Cargar dataset CIFAR-10
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()  # dataset CIFAR-10

# 2. Normalizar imÃ¡genes (0-255 -> 0-1)
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# 3. Convertir labels a categorical (one-hot encoding)
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# 4. Definir nombres de clases
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

print("ğŸ“Š INFORMACIÃ“N DEL DATASET:")
print(f"   ğŸ“ˆ Entrenamiento: {x_train.shape[0]} imÃ¡genes")
print(f"   ğŸ§ª Test: {x_test.shape[0]} imÃ¡genes")
print(f"   ğŸ“ Dimensiones: {x_train.shape[1:]} (HxWxC)")
print(f"   ğŸ“‹ Clases: {num_classes}")

# 5. Configurar batch size para entrenamiento
batch_size = 128
print(f"   ğŸ“¦ Batch size: {batch_size}")

# === CNN SIMPLE DESDE CERO (LIGERAMENTE REGULARIZADA) ===

print("ğŸ—ï¸ IMPLEMENTANDO CNN SIMPLE (casi original + Dropout)")
print("-" * 50)

def create_simple_cnn(input_shape=(32, 32, 3), num_classes=10):
    model = keras.Sequential([
        # Bloque convolucional 1
        layers.Conv2D(32, (3, 3), padding='same', input_shape=input_shape),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        # Bloque convolucional 2
        layers.Conv2D(64, (3, 3), padding='same'),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        # Clasificador
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        # Dropout Ãºnicamente en la parte densa para controlar overfitting
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])

    return model

# Crear modelo CNN simple
simple_cnn = create_simple_cnn()

# Compilar modelo
simple_cnn.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Mostrar arquitectura
simple_cnn.summary()

# Contar parÃ¡metros
total_params = simple_cnn.count_params()
print(f"ğŸ—ï¸ MODELO CNN SIMPLE:")
print(f"   ğŸ”¢ ParÃ¡metros: {total_params:,}")

print("ğŸ¯ IMPLEMENTANDO TRANSFER LEARNING MEJORADO (MobileNetV2)")
print("-" * 50)

def create_transfer_model(input_shape=(32, 32, 3), num_classes=10):
    inputs = keras.Input(shape=input_shape)
    x = layers.Resizing(96, 96)(inputs)
    x = layers.Lambda(preprocess_input)(x)

    base_model = MobileNetV2(
      weights='imagenet',
      include_top=False,
      input_shape=(96, 96, 3),
      name="mobilenetv2_base"
    )

    base_model.trainable = False

    x = base_model(x, training=False)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = keras.Model(inputs, outputs, name="efficientnetb0_transfer")
    return model

# Crear modelo
transfer_model = create_transfer_model()

# Compilar modelo (fase 1: solo cabeza)
transfer_model.compile(
    optimizer=optimizers.Adam(learning_rate=0.0005),  # LR un poco mÃ¡s bajo
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

transfer_model.summary()

total_params = transfer_model.count_params()
trainable_params = sum([tf.keras.backend.count_params(w) for w in transfer_model.trainable_weights])

print("ğŸ¯ MODELO CON TRANSFER LEARNING (FASE 1 - BASE CONGELADA):")
print(f"   ğŸ”¢ ParÃ¡metros totales: {total_params:,}")
print(f"   ğŸ”“ ParÃ¡metros entrenables: {trainable_params:,}")

# 3. Configurar fine-tuning (despuÃ©s de entrenar la cabeza)
def setup_fine_tuning(model, num_unfreeze=20):
    """
    Descongela las Ãºltimas 'num_unfreeze' capas del modelo base
    y recompila con un learning rate mÃ¡s bajo.
    """
    # Recuperar el modelo base por nombre
    base_model = model.get_layer("mobilenetv2_base")
    base_model.trainable = True

    num_layers = len(base_model.layers)
    # Nos aseguramos de no irnos de rango
    num_unfreeze = min(num_unfreeze, num_layers)
    freeze_until = num_layers - num_unfreeze

    # Congelar todas las capas excepto las Ãºltimas `num_unfreeze`
    for layer in base_model.layers[:freeze_until]:
        layer.trainable = False

    # Recompilar con LR mÃ¡s bajo para fine-tuning
    model.compile(
        optimizer=optimizers.Adam(learning_rate=1e-5),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    trainable = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    print(f"ğŸ”“ Fine-tuning configurado - Ãšltimas {num_unfreeze} capas del base entrenables")
    print(f"   ğŸ”“ ParÃ¡metros entrenables: {trainable:,}")

    return model

# === ENTRENAMIENTO DE MODELOS ===

print("ğŸ‹ï¸ CONFIGURANDO ENTRENAMIENTO")
print("-" * 50)

callbacks_list = [
    callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=3,
        restore_best_weights=True
    )
]

# Entrenar CNN simple
print("ğŸ—ï¸ ENTRENANDO CNN SIMPLE...")
simple_history = simple_cnn.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=15,
    validation_data=(x_test, y_test),
    callbacks=callbacks_list,
    verbose=1
)

# Transfer Learning - Fase 1: entrenar solo la cabeza
print("\nğŸ¯ ENTRENANDO TRANSFER LEARNING (FASE 1 - CABEZA)...")
transfer_history_frozen = transfer_model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=10,
    validation_data=(x_test, y_test),
    callbacks=callbacks_list,
    verbose=1
)

# Transfer Learning - Fase 2: Fine-Tuning
print("\nğŸ¯ ACTIVANDO FINE-TUNING EN TRANSFER LEARNING (FASE 2)...")
transfer_model = setup_fine_tuning(transfer_model, num_unfreeze=20)

transfer_history_ft = transfer_model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=10,
    validation_data=(x_test, y_test),
    callbacks=callbacks_list,
    verbose=1
)

# Usamos la historia de la Ãºltima fase para los grÃ¡ficos
transfer_history = transfer_history_ft

# Extraer mÃ©tricas de entrenamiento
simple_train_acc = simple_history.history['accuracy']
simple_test_acc = simple_history.history['val_accuracy']
transfer_train_acc = transfer_history.history['accuracy']
transfer_test_acc = transfer_history.history['val_accuracy']

# === EVALUACIÃ“N FINAL ===

print("ğŸ“Š EVALUACIÃ“N FINAL")
print("-" * 50)

# Evaluar ambos modelos
simple_loss, simple_acc = simple_cnn.evaluate(x_test, y_test, verbose=0)
transfer_loss, transfer_acc = transfer_model.evaluate(x_test, y_test, verbose=0)

print("ğŸ“Š COMPARACIÃ“N FINAL:")
print(f"ğŸ—ï¸ CNN Simple: {simple_acc:.4f} ({simple_acc*100:.2f}%)")
print(f"ğŸ¯ Transfer Learning + Fine-Tuning: {transfer_acc:.4f} ({transfer_acc*100:.2f}%)")
print(f"ğŸ“ˆ Mejora (TL - Simple): {(transfer_acc - simple_acc)*100:+.2f}%")

# Predicciones
simple_predictions = simple_cnn.predict(x_test)
transfer_predictions = transfer_model.predict(x_test)

simple_pred_classes = np.argmax(simple_predictions, axis=1)
transfer_pred_classes = np.argmax(transfer_predictions, axis=1)
true_classes = np.argmax(y_test, axis=1)

# Graficar resultados con anÃ¡lisis de overfitting
plt.figure(figsize=(20, 12))

simple_acc_epochs = range(1, len(simple_train_acc) + 1)
simple_loss_epochs = range(1, len(simple_history.history['loss']) + 1)
transfer_acc_epochs = range(1, len(transfer_train_acc) + 1)
transfer_loss_epochs = range(1, len(transfer_history.history['loss']) + 1)

# 1. PrecisiÃ³n en ValidaciÃ³n
plt.subplot(2, 3, 1)
plt.plot(simple_acc_epochs, simple_test_acc, 'b-', label='CNN Simple', linewidth=2)
plt.plot(transfer_acc_epochs, transfer_test_acc, 'r-', label='Transfer + FT', linewidth=2)
plt.title('PrecisiÃ³n en ValidaciÃ³n', fontsize=14, fontweight='bold')
plt.xlabel('Ã‰poca')
plt.ylabel('PrecisiÃ³n')
plt.legend()
plt.grid(True, alpha=0.3)

# 2. PrecisiÃ³n Final (Bar Chart)
plt.subplot(2, 3, 2)
models = ['CNN Simple', 'Transfer + FT']
accuracies = [simple_acc*100, transfer_acc*100]
bars = plt.bar(models, accuracies, alpha=0.7)
plt.title('PrecisiÃ³n Final', fontsize=14, fontweight='bold')
plt.ylabel('PrecisiÃ³n (%)')
plt.ylim(0, 100)

for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, acc + 1, f'{acc:.1f}%',
             ha='center', fontweight='bold')

# 3. PÃ©rdida - CNN Simple
plt.subplot(2, 3, 3)
plt.plot(simple_loss_epochs, simple_history.history['loss'], 'b-', label='Train', linewidth=2)
plt.plot(simple_loss_epochs, simple_history.history['val_loss'], 'b--', label='Val', linewidth=2)
plt.title('PÃ©rdida - CNN Simple', fontsize=14, fontweight='bold')
plt.xlabel('Ã‰poca')
plt.ylabel('PÃ©rdida')
plt.legend()
plt.grid(True, alpha=0.3)

# 4. PÃ©rdida - Transfer + FT
plt.subplot(2, 3, 4)
plt.plot(transfer_loss_epochs, transfer_history.history['loss'], 'r-', label='Train', linewidth=2)
plt.plot(transfer_loss_epochs, transfer_history.history['val_loss'], 'r--', label='Val', linewidth=2)
plt.title('PÃ©rdida - Transfer Learning + FT', fontsize=14, fontweight='bold')
plt.xlabel('Ã‰poca')
plt.ylabel('PÃ©rdida')
plt.legend()
plt.grid(True, alpha=0.3)

# 5. Overfitting - CNN Simple
plt.subplot(2, 3, 5)
train_acc = simple_history.history['accuracy']
val_acc = simple_history.history['val_accuracy']
plt.plot(simple_acc_epochs, train_acc, 'b-', label='Train', linewidth=2)
plt.plot(simple_acc_epochs, val_acc, 'b--', label='Val', linewidth=2)
plt.title('Overfitting - CNN Simple', fontsize=14, fontweight='bold')
plt.xlabel('Ã‰poca')
plt.ylabel('PrecisiÃ³n')
plt.legend()
plt.grid(True, alpha=0.3)

overfitting_gap_simple = max(train_acc) - max(val_acc)
plt.text(0.02, 0.98, f'Gap: {overfitting_gap_simple:.3f}',
         transform=plt.gca().transAxes, fontsize=10,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))

# 6. Overfitting - Transfer + FT
plt.subplot(2, 3, 6)
train_acc_tl = transfer_history.history['accuracy']
val_acc_tl = transfer_history.history['val_accuracy']
plt.plot(transfer_acc_epochs, train_acc_tl, 'r-', label='Train', linewidth=2)
plt.plot(transfer_acc_epochs, val_acc_tl, 'r--', label='Val', linewidth=2)
plt.title('Overfitting - Transfer Learning + FT', fontsize=14, fontweight='bold')
plt.xlabel('Ã‰poca')
plt.ylabel('PrecisiÃ³n')
plt.legend()
plt.grid(True, alpha=0.3)

overfitting_gap_tl = max(train_acc_tl) - max(val_acc_tl)
plt.text(0.02, 0.98, f'Gap: {overfitting_gap_tl:.3f}',
         transform=plt.gca().transAxes, fontsize=10,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral"))

plt.tight_layout()
plt.show()

print("\nğŸ” ANÃLISIS DE OVERFITTING:")
print(f"ğŸ—ï¸ CNN Simple - Gap Train-Val: {overfitting_gap_simple:.3f}")
print(f"ğŸ¯ Transfer Learning + FT - Gap Train-Val: {overfitting_gap_tl:.3f}")

if overfitting_gap_simple > 0.1:
    print("âš ï¸ CNN Simple muestra overfitting significativo")
if overfitting_gap_tl > 0.1:
    print("âš ï¸ Transfer Learning + FT muestra overfitting significativo")

print("\nğŸ“‹ REPORTE DE CLASIFICACIÃ“N - CNN SIMPLE:")
print(classification_report(true_classes, simple_pred_classes, target_names=class_names))

print("\nğŸ“‹ REPORTE DE CLASIFICACIÃ“N - TRANSFER LEARNING + FINE-TUNING:")
print(classification_report(true_classes, transfer_pred_classes, target_names=class_names))
```
